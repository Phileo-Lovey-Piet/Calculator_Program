```python
def add(x, y):
    return x + y

def sub(x, y):
    return x - y

def mul(x, y):
    return x * y

def div(x, y):
    return x / y

print(f'''
     CALCULATOR
     
     Choose one of the options below by their number.
     
     1. Addition
     2. Subtraction
     3. Multiplication
     4. Division
     
     To exit the program, please type 'exit'.
''')

Selection = input("Enter your selection (either 1,2,3 or 4): ")


while Selection == '1' or Selection == '2' or Selection == '3' or Selection == '4':
    
    numb1 = int(input("Enter first number: "))
    numb2 = int(input("Enter second number: "))
    
    if Selection == '1':
        print(numb1, "+", numb2, "=", add(numb1, numb2))
        
    if Selection == '2':
        print(numb1, "-", numb2, "=", sub(numb1, numb2))
        
    if Selection == '3':
        print(numb1, "x", numb2, "=", mul(numb1, numb2))
        
    if Selection == '4':
        print(numb1, "/", numb2, "=", div(numb1, numb2))
        
        
    Selection = input("Enter your next selection (either 1,2,3 or 4): ")
    

    

    
print(f'''

Calculator turned off.

''')
```
